```python
df = pd.read_excel('/Users/liucheng/Desktop/HPLC/0624_1A01_suc_pyr_glu_ace.xlsx')
df
```

```python
# Filter out rows that start with "MQ"
df_filtered = df[~df['Sample name'].str.startswith('MQ')]
df_filtered
```

```python
def OD_normalized_df (df):
    OD =[0.08,0.105,0.711,0.083, 0.106, 0.726, 0.075, 0.091, 0.754,
     0.063, 0.733, 0.046, 0.754, 0.048, 0.686,
     0.028,0.099, 0.178, 0.439, 0.031, 0.087, 0.164, 0.429,
     0.0695, 0.58,1.015, 0.116, 0.798, 0.958,
     0.17, 0.782, 0.835, 0.679,0.132, 0.786, 0.826, 0.677
    ]
    df_new = df.copy()
    for i in range(len(OD)):
        for col in df_new.columns:
            if col != "Sample name":
                df_new.loc[i,col] = df.loc[i,col]/OD[i]
    return df_new
```

```python
df_filtered = df_filtered.reset_index(drop = True)
df_filtered
```

```python
plot_df = OD_normalized_df(df_filtered)
```

```python
# set the sample name as the x-axis
x =plot_df ['Sample name']
# select the concentration cols as the y-axis
y1 = plot_df['succinate_Conc']
y2 = plot_df['pyruvate_Conc']
y3 = plot_df['acetate_Conc']
y4 = plot_df['glucose_Conc']

plt.figure(figsize = (10,6))
plt.plot(x, y1, marker = 'o', label ='Succinate Concentration')
plt.plot(x, y2, marker='s', label='Pyruvate Concentration')
plt.plot(x, y3, marker='^', label='Acetate Concentration')
plt.plot(x, y4, marker='d', label='Glucose Concentration')

# add labels and title
plt.xlabel('Sample name')
plt.ylabel('Concentration')
plt.title('Concentration of Various Compounds by Sample')
plt.xticks(rotation = 90)

plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages

# Ensure the 'Temperature' column is numeric
df_original['Temperature'] = pd.to_numeric(df_original['Temperature'], errors='coerce')

# Reset the index to ensure it is a range starting from 0
df_original = df_original.reset_index(drop=True)

# List of metabolites to plot
metabolites_glu = ['succinate_Conc', 'pyruvate_Conc', 'acetate_Conc', 'glucose_Conc']
metabolites_glc = [  'acetate_Conc', 'glucose_Conc']
metabolites_glc_special = ['GlcNAc_Conc','succinate_Conc']
letters_glu = ['a', 'b', 'c', 'd']
letters_glc = ['a', 'b']
letters_glc_special =['a', 'b']

# Shape markers for 'glucose' and 'GlcNAc' carbon sources
markers = {'glu': 'o', 'glcnac': 's'}

# Replace negative values in metabolites with NaN
df_original[metabolites_glu] = df_original[metabolites_glu].applymap(lambda x: np.nan if x < 0 else x)
df_original[metabolites_glc] = df_original[metabolites_glc].applymap(lambda x: np.nan if x < 0 else x)
df_original[metabolites_glc_special] = df_original[metabolites_glc_special].applymap(lambda x: np.nan if x < 0 else x)
# specify the file path
file_path_glu ='/Users/liucheng/Desktop/My_thesis_2024/Plots/hplc_glu.pdf'

#create a PdfPages object
with PdfPages(file_path_glu) as pdf:

    # Create two figures with subplots for each carbon source
    fig_hplc_glu, axes_glu = plt.subplots(1, len(metabolites_glu), figsize=(20, 5), sharey=False, constrained_layout=True)
    fig_hplc_glcnac, axes_glcnac = plt.subplots(1, len(metabolites_glc), figsize=(10, 5), sharey=False, constrained_layout=True)
    fig_hplc_glcnac_special, axes_glcnac_special = plt.subplots(1, len(metabolites_glc_special), figsize=(10, 5), sharey=False, constrained_layout=True)
    # Track the maximum y-axis value across all plots for each carbon source
    max_y_values_glu = []
    max_y_values_glcnac = []
    max_y_values_glcnac_special = []
    

    for i, (metabolite_glu, letter_glu) in enumerate(zip(metabolites_glu, letters_glu)):
        # Determine separate y-axis limit for glucose
        if df_original[df_original['Carbon_Source'] == 'glu'][metabolite_glu].max() > 0:
            y_max_glu = df_original[df_original['Carbon_Source'] == 'glu'][metabolite_glu].max()
        else:
            y_max_glu = 1  # if no positive values, set a default limit
        max_y_values_glu.append(y_max_glu)
        # Bar plot for glucose
        sns.barplot(
            data=df_original[df_original['Carbon_Source'] == 'glu'],
            x='Time(h)',
            y=metabolite_glu,
            hue='Temperature',
            palette='coolwarm',
            dodge=True,
            ax=axes_glu[i],
            hue_order=[10, 18,26],  # Order of temperature levels
            edgecolor='black'
        )
        # Set y-axis limits for glucose
        axes_glu[i].set_ylim(0, y_max_glu * 1.1)  # Adjust the multiplier as needed
        # Adding titles and labels for glucose
        axes_glu[i].set_title(f'{metabolite_glu} Concentration (Glucose)')
        axes_glu[i].set_xlabel('Time(h)')
        if i == 0:
            axes_glu[i].set_ylabel(f'Concentration')
        else:
            axes_glu[i].set_ylabel('')
        if i == len(metabolites_glu) - 1:
            handles_glu, labels_glu = axes_glu[i].get_legend_handles_labels()
        axes_glu[i].text(-0.1, 1.05, letter_glu, transform=axes_glu[i].transAxes, 
                         size=20, weight='bold')

        axes_glu[i].legend(handles_glu, labels_glu, title='Temperature', loc='upper right')
    # Adjust layout for glucose
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    fig_hplc_glu.suptitle('Carbon Source: Glucose')
        # Save the figure for glucose to the PDF
    pdf.savefig(fig_hplc_glu)

file_path_glc ='/Users/liucheng/Desktop/My_thesis_2024/Plots/hplc_glc.pdf'        
        
with PdfPages(file_path_glc) as pdf:       

    for i, (metabolite_glc, letter_glc) in enumerate(zip(metabolites_glc, letters_glc)):    
        # Determine separate y-axis limit for GlcNAc
        if df_original[df_original['Carbon_Source'] == 'glcnac'][metabolite_glc].max() > 0:
            y_max_glcnac = df_original[df_original['Carbon_Source'] == 'glcnac'][metabolite_glc].max()
        else:
            y_max_glcnac = 1  # if no positive values, set a default limit
        max_y_values_glcnac.append(y_max_glcnac)
        # Bar plot for GlcNAc
        sns.barplot(
            data=df_original[df_original['Carbon_Source'] == 'glcnac'],
            x='Time(h)',
            y=metabolite_glc,
            hue='Temperature',
            palette='coolwarm',
            dodge=True,
            ax=axes_glcnac[i],
            hue_order=[20, 26],  # Order of temperature levels
            edgecolor='black'
        )

        # Set y-axis limits for GlcNAc
        axes_glcnac[i].set_ylim(0, y_max_glcnac * 1.1)  # Adjust the multiplier as needed

        # Adding titles and labels for GlcNAc
        axes_glcnac[i].set_title(f'{metabolite_glc} Concentration (GlcNAc)')
        axes_glcnac[i].set_xlabel('Time(h)')
        if i == 0:
            axes_glcnac[i].set_ylabel(f'Concentration')
        else:
            axes_glcnac[i].set_ylabel('')
        if i == len(metabolites_glc) - 1:
            handles_glcnac, labels_glcnac = axes_glcnac[i].get_legend_handles_labels()
            axes_glcnac[i].legend(handles_glcnac, labels_glcnac, title='Temperature', loc='upper right')


        # Label the subplot with a letter for GlcNAc
        axes_glcnac[i].text(-0.1, 1.05, letter_glc, transform=axes_glcnac[i].transAxes, 
                            size=20, weight='bold')


    # Adjust layout for GlcNAc
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    fig_hplc_glcnac.suptitle('Carbon Source: GlcNAc')
    pdf.savefig(fig_hplc_glcnac)
    
    
file_path_glc_special ='/Users/liucheng/Desktop/My_thesis_2024/Plots/hplc_glc_special.pdf'     
with PdfPages(file_path_glc_special) as pdf:       

    for i, (metabolite_glc_special, letter_glc_special) in enumerate(zip(metabolites_glc_special, letters_glc_special)):    
        # Determine separate y-axis limit for GlcNAc
        if df_original[df_original['Carbon_Source'] == 'glcnac'][metabolite_glc_special].max() > 0:
            y_max_glcnac_special = df_original[df_original['Carbon_Source'] == 'glcnac'][metabolite_glc_special].max()
        else:
            y_max_glcnac_special = 1  # if no positive values, set a default limit
        max_y_values_glcnac_special.append(y_max_glcnac)
        # Bar plot for GlcNAc
        sns.barplot(
            data=df_original[df_original['Carbon_Source'] == 'glcnac'],
            x='Time(h)',
            y=metabolite_glc_special,
            hue='Temperature',
            palette='coolwarm',
            dodge=True,
            ax=axes_glcnac_special[i],
            hue_order=[20, 26],  # Order of temperature levels
            edgecolor='black'
        )

        # Set y-axis limits for GlcNAc
        axes_glcnac_special[i].set_ylim(0, y_max_glcnac_special * 1.1)  # Adjust the multiplier as needed

        # Adding titles and labels for GlcNAc
        axes_glcnac_special[i].set_title(f'{metabolite_glc_special} Concentration (GlcNAc)')
        axes_glcnac_special[i].set_xlabel('Time(h)')
        if i == 0:
            axes_glcnac[i].set_ylabel(f'Concentration')
        else:
            axes_glcnac[i].set_ylabel('')
        if i == len(metabolites_glc_special) - 1:
            handles_glcnac_special, labels_glcnac_special = axes_glcnac_special[i].get_legend_handles_labels()
            axes_glcnac_special[i].legend(handles_glcnac_special, labels_glcnac_special, title='Temperature', loc='upper right')


        # Label the subplot with a letter for GlcNAc
        axes_glcnac_special[i].text(-0.1, 1.05, letter_glc_special, transform=axes_glcnac_special[i].transAxes, 
                            size=20, weight='bold')


    # Adjust layout for GlcNAc
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    fig_hplc_glcnac_special.suptitle('Carbon Source: GlcNAc')
    pdf.savefig(fig_hplc_glcnac_special)    
# Show the plots
    plt.show()
```
